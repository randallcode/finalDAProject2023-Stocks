```python
#imports the pandas, matplotlib, and seaborn libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reads in the .csv file
stock_financials = pd.read_csv("S&P 500 - Financials - July 2020.csv")

#reads in the .csv file
#the encoding portion of the code is required - without it, a unicode decode error is produced (found the solution on Stack Overflow)
stock_info = pd.read_csv("sp500-companies.csv",encoding='windows-1252')
```

```python
#This code will give us a snapshot of the data in the financial dataframe
stock_financials.head()
```

```python
#This code will give us a snapshot of the data in the information dataframe
stock_info.head()
```

```python
#This code will provide the total rows & columns in the stockfinancial df.
stock_financials
```

```python
#This code will provide the total rows & columns in the stockinfo df.
stock_info
```

```python
#Renaming the "Ticker" column to match the column name in the stocks_financials df
stock_info.rename(columns={"Ticker" : "Symbol"}, inplace="False")

#verifying that the name actually changed
stock_info.head()
```

```python
#This delete the "Name" column from the stock_info df - without doing this, there will two "name" columns in the stocks_merged df below, which is what we don't want
stock_info = stock_info.drop("Name", axis="columns")
```

```python
#This deletes the "SEC Filings" column from the stock_financial df
stock_financials = stock_financials.drop("SEC Filings", axis="columns")
```

```python
#merges the two dataframes together based upon the Symbol column
stocks_merged = pd.merge(stock_financials,stock_info, on="Symbol", how="outer")
```

```python
#displays information behind the newly merged df
stocks_merged
```

```python
#this provides us various quantitative data for each column that uses numerical values
stocks_merged.describe()
```

```python
#this just shows some information behind the stocks_merged df
stocks_merged.info()
```

```python
#This sorts the stocks_merged df by the "Price" column by the highest price
stocks_merged_sorted = stocks_merged.sort_values(by="Price", ascending=False)
stocks_merged_sorted
```

```python
#This sorts the data to the top 5 companies by price
top_five_companies = stocks_merged_sorted.head(6).dropna()
top_five_companies
```

```python
#creates a bar chart of the top 5 companies based upon price
plt.bar(top_five_companies["Symbol"],top_five_companies["Price"],edgecolor="black")
plt.xlabel("Symbol")
plt.ylabel("Price $")
plt.title("Top 5 Companies Based Upon Their Stock Price")

plt.show()
```

```python
#Lists off the unique values in the "Sector" column
stocks_merged_unique = stocks_merged["Sector"].unique()
print(stocks_merged_unique)
```

```python
#Creates a pie chart that removes two sectors to make it easier to read
excluded_categories = ["Materials", "Telecommunication Services"]
stocks_merged_filtered = stocks_merged[~stocks_merged["Sector"].isin(excluded_categories)]
stocks_merged_filtered['Sector'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Sectors in S&P 500')
plt.ylabel('')  # Remove the default 'Sector' label on the y-axis
plt.show()
```

```python
#Lists the 10 oldest companies based upon their founded year
oldest_companies = stocks_merged.sort_values(by= "Founded",ascending=True)
oldest_companies.head(10)
```

```python
#This returns the companies that were founded in 1995 or later, and that have a Price/Earnings value greater than 70
stocks_merged_founded = stocks_merged[(stocks_merged["Founded"] >= "1995") & (stocks_merged["Price/Earnings"] >= 70)]
stocks_merged_founded
```

```python
#This returns the companies that were founded in 2015 or later, while droping null values
stocks_founded = stocks_merged[(stocks_merged["Founded"] >= "2015")]
stocks_founded.dropna()
```
