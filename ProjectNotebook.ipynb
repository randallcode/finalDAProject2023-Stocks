```python
#imports the pandas library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reads in the .csv file
stock_financials = pd.read_csv("S&P 500 - Financials - July 2020.csv")

#reads in the .csv file
#the encoding portion of the code is required - without it, a unicode decode error is produced (found the solution on Stack Overflow)
stock_info = pd.read_csv("sp500-companies.csv",encoding='windows-1252')
```

```python
#Renaming the "Ticker" column to match the column name in the stocks_financials df
stock_info.rename(columns={"Ticker" : "Symbol"}, inplace="False")

#verifying that the name actually changed
stock_info.head()
```

```python
#merges the two dataframes together based upon the Symbol column
stocks_merged = pd.merge(stock_financials,stock_info, on="Symbol", how="outer")
```

```python
#displays information behind the newly merged df
stocks_merged
```
