```python
#imports the pandas and matplotlib libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reads in the .csv file
stock_financials = pd.read_csv("S&P 500 - Financials - July 2020.csv")

#reads in the .csv file
#the encoding portion of the code is required - without it, a unicode decode error is produced (found the solution on Stack Overflow)
stock_info = pd.read_csv("sp500-companies.csv",encoding='windows-1252')
```

```python
#Renaming the "Ticker" column to match the column name in the stocks_financials df
stock_info.rename(columns={"Ticker" : "Symbol"}, inplace="False")

#verifying that the name actually changed
stock_info.head()
```

```python
#This delete the "Name" column from the stock_info df - without doing this, there will two "name" columns in the stocks_merged df below, which is what we don't want
stock_info = stock_info.drop("Name", axis="columns")
```

```python
#This deletes the "SEC Filings" column from the stock_financial df
stock_financials = stock_financials.drop("SEC Filings", axis="columns")
```

```python
#merges the two dataframes together based upon the Symbol column
stocks_merged = pd.merge(stock_financials,stock_info, on="Symbol", how="outer")
```

```python
#displays information behind the newly merged df
stocks_merged
```

```python
#this provides us various quantitative data for each column that uses numerical values
stocks_merged.describe()
```

```python
#this just shows some information behind the stocks_merged df
stocks_merged.info()
```

```python
#Printing the df just to be able to view some of the data without scrolling up
stocks_merged
```

```python
#This sorts the stocks_merged df by the "Price" column by the highest price
stocks_merged_sorted = stocks_merged.sort_values(by="Price", ascending=False)
stocks_merged_sorted
```

```python
#This sorts the data to the top 5 companies by price
top_five_companies = stocks_merged_sorted.head(5)
top_five_companies
```

```python
#creates a bar chart of the top 5 companies based upon price
plt.bar(top_five_companies["Symbol"],top_five_companies["Price"])
plt.xlabel("Symbol")
plt.ylabel("Price")
plt.title("Top 5 Companies and Their Stock Price")

plt.show()
```
